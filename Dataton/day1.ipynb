```python
condition = deliver["order_status"] != "delivered"
deliver.loc[condition, :]
```

```python
delivered_order = deliver.loc[~condition, :]
delivered_order.head()
```

```python
delivered_order[["order_delivered_customer_date", "order_estimated_delivery_date"]]
```

```python
real_date = delivered_order["order_delivered_customer_date"]
real_date.head()
```

```python
est_date = delivered_order["order_estimated_delivery_date"]
est_date.head()
```

```python
real_date = pd.to_datetime(real_date)
est_date = pd.to_datetime(est_date)
```

```python
diff_date = est_date - real_date
diff_date = diff_date.astype("str")
diff_date
```

```python
diff_date = diff_date.map(lambda x: x.split()[0])
condition = diff_date != "NaT"

diff_date = diff_date[condition]

diff_date = diff_date.map(lambda x: int(x) * -1 if int(x) < 0 else x)
diff_date.head()
```

```python
diff_date = diff_date.astype("int")
diff_date.describe()
```

```python
diff_date.value_counts().plot().set_title("도착 예상일과 실제 도착일 차이")
```
